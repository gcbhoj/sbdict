```python
from bs4 import BeautifulSoup
import requests
import urllib3
import json

urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

url = "https://learnsanskrit.cc/fables/getFables"

page = requests.get(url, verify=False)

# Parse HTML instead of trying to read JSON
soup = BeautifulSoup(page.text, "html.parser")

# Example: print formatted HTML
print(soup.prettify())

# If you want structured extraction instead:
data = {
    "title": soup.title.string if soup.title else None
}

print(json.dumps(data, indent=4))
```

```python
img_tags = soup.find_all("img")
print (img_tags)
```

```python
jpg_names = []

for img in img_tags:
    src = img.get("src")

    if src:
        # clean bad characters
        src = src.replace('\\"', '').replace('"', '').strip().strip(",")

        # get filename only
        filename = src.split("/")[-1]

        if filename.lower().endswith(".jpg"):
            jpg_names.append(filename)

print(jpg_names)
```

```python
cleaned_vendor_id = [name.replace("_icon.jpg","") for name in jpg_names]
print(cleaned_vendor_id)
```

```python
story_description = soup.find_all(
    class_='\\"fabdeclaration\\"'
)
print(story_description)
```

```python
results = []

for item in story_description:

    title_tag = item.find(["b", "strong"])

    title = title_tag.get_text(strip=True) if title_tag else "Unknown"

    # remove title tag temporarily
    if title_tag:
        title_tag.extract()

    description = item.get_text(" ", strip=True)

    results.append({
        "title": title,
        "description": description
    })

print(results)
```

```python
from uuid import uuid4
from datetime import datetime
import json

# final structured result
final_result = []

# -----------------------------------
# STEP 1: CREATE CATEGORY STRUCTURE
# -----------------------------------

category_map = {}

for item in results:

    title = item["title"].strip()

    category_map[title] = {
        "_id": str(uuid4()),
        "title": title,
        "description": item["description"],
        "origin": title,
        "story_description": []
    }

# -----------------------------------
# STEP 2: ADD STORIES TO CATEGORY
# -----------------------------------

for vendor_id, page_data in zip(cleaned_vendor_id, complete_page):

    data = page_data.get("data", {})

    author = data.get("author", "").strip()

    story_title = (
        data.get("summary_head", [""])[0]
        if data.get("summary_head")
        else ""
    )

    story_object = {
        "_id": str(uuid4()),
        "vendorId": vendor_id,
        "storyTitle": story_title,
        "used": False,
        "updatedOn": datetime.utcnow().isoformat()
    }

    # attach story to correct category
    if author in category_map:
        category_map[author]["story_description"].append(
            story_object
        )

# -----------------------------------
# STEP 3: CONVERT TO ARRAY
# -----------------------------------

final_result = list(category_map.values())

# -----------------------------------
# OUTPUT
# -----------------------------------

print(json.dumps(final_result, indent=4))
```
