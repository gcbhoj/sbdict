```python
from utils.file_system_reader import FileSystemReader
```

```python
FILE_NAME = "stories.json"
```

```python
read_req = FileSystemReader(FILE_NAME)
MAIN_DATA = read_req.read_file()
print(MAIN_DATA)
```

```python
sanskrit_version = MAIN_DATA.get("sanskritVersion")
print(sanskrit_version)
```

```python
cleaned_words =[]

for sentence in sanskrit_version:
    for word in sentence.split():
        cleaned_words.append(word)

print(cleaned_words)
    
```

```python
import stanza
nlp = stanza.Pipeline("sa",processors='tokenize,pos,lemma')
```

```python
tokened_words = []

for word in cleaned_words:

    # word is string
    doc = nlp(word)

    # processed tokens are inside doc
    for sent in doc.sentences:
        for token in sent.words:

            tokened_words.append({
                "text": token.text,
                "lemma": (
                    token.lemma.strip("-").split("_")[0]
                    if token.lemma
                    else token.text
                ),
                "upos": token.upos,
                "xpos": token.xpos if token.xpos else None,
                "feats": token.feats if token.feats else "_"
            })

print(tokened_words)
```

```python
MAIN_DATA["tokenized_sanskrit_version"] = tokened_words
```
